```python
os.makedirs("./submit_log", exist_ok=True)
for task_id in task_list:
    data_list_list = bench_data[task_id]["data_list"]
    query = bench_data[task_id]["goal_description"]
    for rep_id, data_list in enumerate(data_list_list):
        log_path = f'./submit_log/{task_id}_rep{rep_id+1}.json'
        if os.path.exists(log_path):
            print(f"Skip {task_id}-rep{rep_id+1}: log file already exists.")
            continue
        output_dir = create_output_dir(user_id="analyst_agent", task=f"{task_id}_rep{rep_id+1}")
        try:
            submit_task = await retrieve_plan_submit(
                goal_description=query,
                data_list=data_list,
                user_id="analyst_agent",
                is_create_dir=False,
                output_dir=output_dir
            )
        except Exception as e:
            print(e)
            print(f"Submit task {task_id}-rep{rep_id+1} failed.")
            continue
        with open(log_path, 'w') as fo:
            json.dump(submit_task, fo, indent=4)
        time.sleep(30)
```
