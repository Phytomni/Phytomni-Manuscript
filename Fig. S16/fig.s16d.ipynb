```python
import matplotlib.pyplot as plt
```

## Fig S1D

```python
data_list = {
    'Genome': 26, 
    'Transcript': 29, 
    'Metabolome': 1, 
    'Metagenome': 20, 
    'Protein': 16, 
    'Structure': 4, 
    'Epigenetics': 24, 
    'Environment': 9, 
    'Phenomics': 2, 
    'Phylogeny': 22, 
    'Single-Cell': 4, 
    'Spatial': 10
}
colors = ['#CC625F', '#FCECCA', '#FFD377', '#C0E56F', '#FFFB79', '#A0FCE8', '#FEC1E2', '#D4C1DC', '#D7D7D5', '#8EB4D0', '#F49513', '#FCC3B4']
# data
categories = data_list.keys()
sales = data_list.values()
explode = [0.1] * len(categories)
```

```python
fig, ax = plt.subplots(figsize=(10, 7), dpi=300)
# Plot
wedges, texts, autotexts = ax.pie(
    sales, 
    explode=explode,
    colors=colors,
    autopct=lambda p: f'{p * 167 / 100:.0f}',
    pctdistance=0.75,
    textprops={'fontsize': 10, 'color': 'black'},
    wedgeprops={
        'edgecolor': 'black',
        'linewidth': 1,
        'linestyle': '-',
        'antialiased': True
    }
)

ax.legend(
    wedges, 
    categories,
    title="Tools Classification",
    loc="center left",
    bbox_to_anchor=(0.85, 0.25, 1, 1)
)

plt.title('Agents Tools Summary', fontsize=16, pad=20)

ax.axis('equal')  

plt.setp(autotexts, size=10, weight='bold', color='black')
plt.tight_layout()
plt.show()
# plt.savefig('figs1d.pdf')
```
