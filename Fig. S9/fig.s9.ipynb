```python
n_groups = len(data)
index = np.arange(len(categories))
bar_width = 0.10
opacity = 1

fig, ax = plt.subplots(figsize=(16, 8), dpi=300)

def autolabel(rects):
    """"""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:.2f}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')

for i, (sample, values) in enumerate(data.items()):
    rects = ax.bar(index + i * bar_width, values, bar_width, label=sample, alpha=opacity, color=my_colors[sample], edgecolor='black', linewidth=1)

ax.set_ylabel('Task Exceution Rate(%)', fontsize=15)
ax.set_title('Exceution Ability(Phyto Models)', fontsize=20)
ax.set_xticks(index + bar_width * (len(data) - 1) / 2)
ax.set_xticklabels(categories, fontsize=15, rotation=90)
ax.legend(fontsize=15)
plt.yticks(fontsize=15, ha='right')
plt.ylim(0, 100)
plt.tick_params(
    axis='both',
    which='major',
    width=1,
    length=8 
)

plt.tight_layout()
plt.show()
# plt.savefig('figs9a.pdf')
```
