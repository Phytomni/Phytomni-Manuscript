```python
data = np.array(r_list)
normalized_data = np.zeros_like(data)
for col in range(data.shape[1]):
    col_data = data[:, col]
    min_val = col_data.min()
    max_val = col_data.max()
    if max_val == min_val:
        normalized_data[:, col] = 0.5
    else:
        normalized_data[:, col] = (col_data - min_val) / (max_val - min_val)
for i, metric in enumerate(theta_list):
    col_min = normalized_data[:, i].min()
    col_max = normalized_data[:, i].max()
    print(f"{metric}: [{col_min:.3f}, {col_max:.3f}]")
normalized_r_list = normalized_data.tolist()
```

```python
line_width = 2

fig = go.Figure()

for i in range(8):
    fig.add_trace(go.Scatterpolar(
        r=normalized_r_list[i]+[normalized_r_list[i][0]],
        theta=theta_list+[theta_list[0]],
        name=model_list[i],
        mode='lines',
        line_color=color_list[i],
        line_width=line_width,
        opacity=0.5,
        fill = 'toself',
    ))
fig.update_layout(
    template=None,
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[0, 1],
            linewidth=line_width)),
    showlegend=True,
    legend=dict(orientation = 'h',),
    plot_bgcolor='white',
    font_family='Arial',
    font_color='black',
    font_size=20,
    width=1920,
    height=1080)

fig.show()
# file_prefix = f'supplementary_fig.13.phytobench-review.polar'.lower().replace(' ', '_')
# fig.write_image(f'{file_prefix}.pdf')
# fig.write_image(f'{file_prefix}.png')
```
