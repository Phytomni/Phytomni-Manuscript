### Fig. 3e

```python
df = pd.read_excel('PhytoBench-Paper-for_plot.xlsx')
df_filtered = df[(df['Task category'].notna()) & (df['Check'] == 'Y')]
df_filtered['Task category'] = df_filtered['Task category'].str.strip().str.capitalize()
df_filtered.loc[df_filtered['Task category'] == 'Data visualization', 'Task category'] = 'Visualization'
task_category_counts = df_filtered['Task category'].value_counts()
```

```python
p_task_time_rate, h1_task_time_rate, h2_task_time_rate = [], [], []
for task_category in ['Code execution', 'Visualization', 'Tool installation', 'Result analysis', 'Format conversion', 'Data download']:
    time_rate = []
    for p_time in ['1h', '3h', '6h', '12h', '18h', '24h']:
        df_task_time = df_filtered[df_filtered['Task category']==task_category][p_time]
        time_rate.append(np.sum(df_task_time)/df_task_time.shape[0])
    p_task_time_rate.append(time_rate)
    time_rate = []
    for h1_time in ['1h.1', '3h.1', '6h.1', '12h.1', '18h.1', '24h.1']:
        df_task_time = df_filtered[df_filtered['Task category']==task_category][h1_time]
        time_rate.append(np.sum(df_task_time)/df_task_time.shape[0])
    h1_task_time_rate.append(time_rate)
    time_rate = []
    for h2_time in ['1h.2', '3h.2', '6h.2', '12h.2', '18h.2', '24h.2']:
        df_task_time = df_filtered[df_filtered['Task category']==task_category][h2_time]
        time_rate.append(np.sum(df_task_time)/df_task_time.shape[0])
    h2_task_time_rate.append(time_rate)
```

```python
p_task_time_rate_array = np.array(p_task_time_rate)
p_task_time_rate_array[:, 1] = p_task_time_rate_array[:, 0] + p_task_time_rate_array[:, 1]
p_task_time_rate_array[:, 2] = p_task_time_rate_array[:, 1] + p_task_time_rate_array[:, 2]
p_task_time_rate_array[:, 3] = p_task_time_rate_array[:, 2] + p_task_time_rate_array[:, 3]
p_task_time_rate_array[:, 4] = p_task_time_rate_array[:, 3] + p_task_time_rate_array[:, 4]
p_task_time_rate_array[:, 5] = p_task_time_rate_array[:, 4] + p_task_time_rate_array[:, 5]

h_task_time_rate_array = (np.array(h1_task_time_rate) + np.array(h2_task_time_rate))/2
h_task_time_rate_array[:, 1] = h_task_time_rate_array[:, 0] + h_task_time_rate_array[:, 1]
h_task_time_rate_array[:, 2] = h_task_time_rate_array[:, 1] + h_task_time_rate_array[:, 2]
h_task_time_rate_array[:, 3] = h_task_time_rate_array[:, 2] + h_task_time_rate_array[:, 3]
h_task_time_rate_array[:, 4] = h_task_time_rate_array[:, 3] + h_task_time_rate_array[:, 4]
h_task_time_rate_array[:, 5] = h_task_time_rate_array[:, 4] + h_task_time_rate_array[:, 5]

shift_array = (p_task_time_rate_array - h_task_time_rate_array)

shift_array_max = np.max(shift_array)
shift_array_min = np.min(shift_array)
shift_max_min = shift_array_max - shift_array_min
```

```python
line_width = 2
fig = go.Figure()

fig.add_trace(go.Heatmap(
    x=['1h', '3h', '6h', '12h', '18h', '24h'],
    y=['Data download', 'Tool installation', 'Code execution', 'Format conversion', 'Result analysis', 'Visualization'],
    z=shift_array[[5, 2, 0, 4, 3, 1]],
    colorscale = [
        [0, cl.flipper()['seq']['5']['Oranges'][4]],
        [(shift_array_min*4/5-shift_array_min)/shift_max_min, cl.flipper()['seq']['5']['Oranges'][3]],
        [(shift_array_min*3/5-shift_array_min)/shift_max_min, cl.flipper()['seq']['5']['Oranges'][2]],
        [(shift_array_min*2/5-shift_array_min)/shift_max_min, cl.flipper()['seq']['5']['Oranges'][1]],
        [(shift_array_min*1/5-shift_array_min)/shift_max_min, cl.flipper()['seq']['5']['Oranges'][0]],
        [(0-shift_array_min)/shift_max_min, 'rgb(255,255,255)'],
        [(shift_array_max*1/5-shift_array_min)/shift_max_min, cl.flipper()['seq']['5']['Blues'][0]],
        [(shift_array_max*2/5-shift_array_min)/shift_max_min, cl.flipper()['seq']['5']['Blues'][1]],
        [(shift_array_max*3/5-shift_array_min)/shift_max_min, cl.flipper()['seq']['5']['Blues'][2]],
        [(shift_array_max*4/5-shift_array_min)/shift_max_min, cl.flipper()['seq']['5']['Blues'][3]],
        [1, cl.flipper()['seq']['5']['Blues'][4]]
    ],
))

fig.update_layout(
    showlegend=False,
    yaxis_scaleanchor='x',
    xaxis={'showline': False, 'linewidth': line_width, 'linecolor': 'black', 'mirror': False, 'ticks': 'outside', 'tickwidth': line_width},
    yaxis={'showline': False, 'linewidth': line_width, 'linecolor': 'black', 'mirror': False, 'ticks': 'outside', 'tickwidth': line_width},
    plot_bgcolor='white',
    font_family='Arial',
    font_color='rgb(0,0,0)',
    font_size=20,
    width=1080,
    height=1080)

fig.show()
# file_prefix = f'fig.3e.phytobench-paper.heatmap'.lower().replace(' ', '_')
# fig.write_image(f'{file_prefix}.pdf')
# fig.write_image(f'{file_prefix}.png')
```
