```python
import os
SAVE_FIGS = os.getenv("PHYTOMNI_SAVE") == "1"
if SAVE_FIGS:
    os.makedirs("output", exist_ok=True)
```

```python
from json import load

import numpy as np
import pandas as pd
import plotly.graph_objects as go
import plotly.io as pio
import plotly.offline as offline
```

```python
offline.init_notebook_mode(connected=True)
pio.kaleido.scope.default_format = 'pdf'
```

```python
with open('Model-Loss-for_plot/Phyto-Chatbot-Pretrain.loss.json', 'r') as open_json:
    chat_pretrain_loss = load(open_json)
chat_pretrain_loss_y = [l['loss'] for l in chat_pretrain_loss['loss']]
```

```python
with open('Model-Loss-for_plot/Phyto-Reasoner-Pretrain.loss.json', 'r') as open_json:
    reasoner_pretrain_loss = load(open_json)
reasoner_pretrain_loss_y = [l['loss'] for l in reasoner_pretrain_loss['loss']]
```

```python
line_width = 2

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=list(range(len(chat_pretrain_loss_y))),
    y=chat_pretrain_loss_y,
    line_width=line_width,
    name='Phyto-Chatbot',
))

fig.add_trace(go.Scatter(
    x=list(range(len(reasoner_pretrain_loss_y))),
    y=reasoner_pretrain_loss_y,
    line_width=line_width,
    name='Phyto-Reasoner',
))

fig.update_layout(
    xaxis={'showline': True, 'linewidth': line_width, 'linecolor': 'rgb(0,0,0)', 'mirror': True, 'ticks': 'outside', 'tickwidth': line_width, 'range': [0, 1060]},
    yaxis={'showline': True, 'linewidth': line_width, 'linecolor': 'rgb(0,0,0)', 'mirror': True, 'ticks': 'outside', 'tickwidth': line_width, 'range': [1.7, 2]},
    plot_bgcolor='white',
    font_family='Arial',
    font_color='rgb(0,0,0)',
    font_size=20,
    width=1920,
    height=1080/2)
fig.update_traces(mode='lines+markers')

fig.show()
if SAVE_FIGS:
    file_prefix = f'supplementary_fig.1a.model_pretrain_loss.line'.lower().replace(' ', '_')
    fig.write_image(f"output/{file_prefix}.pdf")
    fig.write_image(f"output/{file_prefix}.png")
```
