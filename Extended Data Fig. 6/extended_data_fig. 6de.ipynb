```python
# %% plot GeneTuring
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap

%matplotlib inline

data_dict = {
    'Models': [
        'GPT-5', 'o3', 'Gemini-2.5-Pro', 'Claude-4.1', 
        'Grok-3-Beta', 'DeepSeek-V3', 'DeepSeek-R1', 'GeneGPT', 'Phyto-Reasoner'
    ],
    'Gene alias': [0.28, 0.28, 0.18, 0.24, 0.32, 0.18, 0.2, 0.8, 0.96],
    'Gene name conversion': [0, 0, 0, 0.14, 0.1, 0.02, 0.02, 1, 1],
    'Gene location': [0, 0.34, 0.3, 0.44, 0.46, 0.32, 0.36, 0.62, 0.94],
    'SNP location': [0, 0.04, 0.34, 0, 0.02, 0.02, 0.02, 1, 1],
    'Gene SNP association': [0, 0, 0, 0, 0, 0, 0, 1, 1],
    'Protein−coding genes': [0.9, 0.84, 0.94, 0.78, 0.98, 0.88, 0.96, 0.76, 1],
    'Human genome DNA aligment': [0, 0, 0, 0, 0, 0, 0, 0.44, 0.98],
    'Multi−species DNA aligment': [0, 0.02, 0, 0, 0, 0.08, 0.1, 0.86, 1]
}

data = pd.DataFrame(data_dict)
data.set_index("Models", inplace=True)

colors = ["#ffffff", "#A8ABE0"]
gradient_cmap = LinearSegmentedColormap.from_list("gradient", colors)


annot_data = data.copy()

annot_data = annot_data.applymap(lambda x: "NA" if x == 0 else f"{x:.2f}")

plt.figure(figsize=(10, 8))
sns.heatmap(
    data, 
    annot=annot_data,      
    cmap=gradient_cmap, 
    vmin=0, 
    vmax=1, 
    fmt="",                
    cbar_kws={"label": "Accuracy"} 
)
plt.title("Accuracy Assessment of GeneTuring", fontsize=14, pad=15)
plt.xticks(rotation=45, ha="right")
plt.xlabel("")
plt.ylabel("")

plt.tight_layout()
plt.show()
plt.close()
```
